```python
import requests
url='https://finance.naver.com/marketindex/?tabSel=exchange#tab_section'
res=requests.get(url)
```

```python
c=input('나라를 입력하여 주십시오.')
m=int(input('원하는 금액을 입력하여 주십시오.'))
from bs4 import BeautifulSoup
markup=res.text
parser='lxml'
soup=BeautifulSoup(markup,parser)
c1=soup.find_all('option')
print(c1)
import re
num=re.findall('\d+.\d+(?=">)',str(c1))
print(num)
cont=re.findall('[가-힣]+(?= [가-힣]+)',str(c1))
print(cont)
c2=c1.index(c)
print(num[c2-1]*m)
```
